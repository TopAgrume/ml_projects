## R√©sum√© statistique personnalis√© des donn√©es <a class="anchor" id="32"></a>

Dans cette partie, on essaye de r√©cup√©rer une id√©e globales des valeurs possibles pour chaque feature.
```python
import matplotlib.pyplot as plt

def get_summary(df):
    df_desc = pd.DataFrame(df.describe(include='all').T)
    df_summary = pd.DataFrame({
        'dtype': df.dtypes,
        'unique':df.nunique().values,
        'missing': df.isna().sum().values,
        'duplicates': df.duplicated().sum(),
        'min': df_desc['min'].values,
        'max': df_desc['max'].values,
        'avg': df_desc['mean'].values,
        'std dev': df_desc['std'].values
    })
    return df_summary

get_summary(data_df).style.background_gradient(cmap='viridis_r',low=0.8)
```

# Step 2: Data Preprocessing <a class="anchor" id="4"></a>

On applique une s√©paration entre les features d'entrainement et les labels de r√©f√©rences qui nous servirons pour l'entrainement.

```python
# Define features and target labels
X = data_df.drop('Diagnosis', axis=1)
y = data_df['Diagnosis']

print(X.shape)
print(y.shape)
```

## Prise en main des valeurs manquantes et Normalisation

Notre but va maintenant √™tre de normaliser les donn√©es continues entre 0 et 1. Comme nous avons pu voir dans les pr√©c√©dentes analyses, certaines features prennent leurs valeurs dans des intervalles qui ont une borne inf√©rieur sup√©rieure √† 0. C'est pour cette raison que nous avons choisi d'appliquer un `MinMaxScaler`.

```python
#from sklearn.impute import SimpleImputer
#from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

min_max_scaling = Pipeline(steps=[
    ('scaler', MinMaxScaler())
])

# Apply preprocessing
X_preprocessed = min_max_scaling.fit_transform(X)
```

# Step 3: Selection des mod√®les et entrainement <a class="anchor" id="5"></a>

Pour la classification, on a d√©cid√© de s'appuyer sur 3 mod√®les, √† savoir:

- La r√©gression logistique
- La classification par random forest
- Le classification par vecteurs de support

Afin d'obtenir une √©valuation plus robuste et fiable des performances du mod√®le, nous avons mis en place l'utilisation de la validation crois√©e par les fonctions d'√©tudes Optuna.
Cela nous aide √† s√©lectionner des mod√®les et des hyperparam√®tres qui offriront les meilleures performances lors de l'application sur de nouvelles donn√©es.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Objective function for Logistic Regression
def objective_lr(trial):
    param = {
        'solver': trial.suggest_categorical('solver', ['liblinear', 'saga']),
        'C': trial.suggest_float('C', 1e-4, 1e2, log=True),
        'penalty': trial.suggest_categorical('penalty', ['l2', 'l1']),
    }
    # We set the solver to
    model = LogisticRegression(max_iter=10000, **param)
    score = cross_val_score(model, X_preprocessed, y, cv=5, scoring='accuracy').mean()
    return score

# Objective function for Random Forest
def objective_rf(trial):
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 300),
        'max_depth': trial.suggest_int('max_depth', 2, 32, log=True),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 16),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 16)
    }
    model = RandomForestClassifier(**param)
    score = cross_val_score(model, X_preprocessed, y, cv=5, scoring='accuracy').mean()
    return score

# Objective function for SVM
def objective_svm(trial):
    param = {
        'C': trial.suggest_float('C', 1e-4, 1e2, log=True),
        'gamma': trial.suggest_float('gamma', 1e-4, 1e-1, log=True),
        'kernel': trial.suggest_categorical('kernel', ['rbf', 'linear'])
    }
    model = SVC(**param)
    score = cross_val_score(model, X_preprocessed, y, cv=5, scoring='accuracy').mean()
    return score
```

```python
import optuna

study_lr = optuna.create_study(direction='maximize')
%time study_lr.optimize(objective_lr, n_trials=50)
```

```python
study_rf = optuna.create_study(direction='maximize')
%time study_rf.optimize(objective_rf, n_trials=50)
```

```python
study_svm = optuna.create_study(direction='maximize')
%time study_svm.optimize(objective_svm, n_trials=50)
```

On r√©cup√®re les meilleurs param√®tres de chaque mod√®le.

```python
# Best parameters for each model
best_params_lr = study_lr.best_params
best_params_rf = study_rf.best_params
best_params_svm = study_svm.best_params

print("Best parameters for Logistic Regression:", best_params_lr)
print("Best parameters for Random Forest:", best_params_rf)
print("Best parameters for SVM:", best_params_svm)
```

# Step 4: Evaluation des mod√®les <a class="anchor" id="6"></a>

On va maintenant r√©entrainer les mod√®les avec leurs meilleurs hyperparam√®tres et les √©valuer plus en d√©tails en s√©parant bien `avec` et `sans` preprocessing.

```python
from sklearn.model_selection import train_test_split

# Without Preprocessing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# With Preprocessing
X_preprocessed_train, X_preprocessed_test, y_preprocessed_train, y_preprocessed_test = train_test_split(X_preprocessed, y, test_size=0.3, random_state=42)

print("Training dataset length:", len(X_train))
print("Test dataset length:", len(X_test))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

def compute_visual_and_statistical_report(name, y_ref, y_predicition, training_time, proceprocessing=False):
    cm = confusion_matrix(y_ref, y_predicition)
    cr = classification_report(y_ref, y_predicition)

    # Display results (confusion_matrix, and classification_report)
    print(f"Model: {name} with", f"{proceprocessing=}")
    print(f'Time taken: {training_time:4f} seconds')
    print("Classification Report:")
    print(cr)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
models_without_preprocessing = {
    'Logistic Regression': LogisticRegression(max_iter=10000, **best_params_lr),
    'Random Forest': RandomForestClassifier(**best_params_rf),
    'SVM': SVC(**best_params_svm)
}

models_with_preprocessing = {
    'Logistic Regression': LogisticRegression(max_iter=10000, **best_params_lr),
    'Random Forest': RandomForestClassifier(**best_params_rf),
    'SVM': SVC(**best_params_svm)
}

for name, model in models_without_preprocessing.items():
    training_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - training_time
    y_pred = model.predict(X_test)

    compute_visual_and_statistical_report(name, y_test, y_pred, training_time)

for name, model in models_with_preprocessing.items():
    training_time = time.time()
    model.fit(X_preprocessed_train, y_preprocessed_train)
    training_time = time.time() - training_time
    y_preprocessed_pred = model.predict(X_preprocessed_test)

    compute_visual_and_statistical_report(name, y_preprocessed_test, y_preprocessed_pred, training_time, proceprocessing=True)
```

Pour notre √©tude, nous avons donc les r√©sultats suivants avec les meilleurs hyperparam√®tres trouv√© sur 50 essais:

**Sans Pr√©traitement des features:**

| Mod√®le              | Accuracy | F1 Score | Precision | Recall | Temps d'ex√©cution (s) |
|---------------------|----------|----------|-----------|--------|-----------------------|
| Logistic Regression | 0.80     | 0.80     | 0.80      | 0.80   | 2.507859              |
| Random Forest       | **0.91** | **0.91** | **0.91**  |**0.91**| **0.388567**          |
| SVM                 | 0.81     | 0.81     | 0.81      | 0.81   | 125.423846            |

**Avec Pr√©traitement des features:**

| Mod√®le              | F1 Score | Precision | Recall | Temps d'ex√©cution (s) |
|---------------------|----------|-----------|--------|-----------------------|
| Logistic Regression | 0.82     | 0.82      | 0.82   | **0.009628**          |
| Random Forest       | **0.90** | **0.90**  |**0.90**| 0.430003              |

Pour cette √©tude, la Random Forest est le mod√®le √† privil√©gier en raison de ses performances √©lev√©es et de son temps d'entrainement raisonnable, ind√©pendamment de l'utilisation du pr√©traitement. Le pr√©traitement n'a pas montr√© d'am√©lioration significative sur la random forest, ni sur le mod√®le SVM. Seule la r√©gression logistique a gagn√© en score (~ 2% suppl√©mentaires). Le pr√©processing peut √™tre √©vit√© (ou alors modifi√©) pour simplifier le processus et r√©duire le temps de calcul.
